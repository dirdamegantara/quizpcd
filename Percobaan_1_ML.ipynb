```python
import string
import nltk
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

nltk.download('punkt')
nltk.download('stopwords')
```

```python
text = "Perkenalkan nama saya Dirda Megantara, saya adalah mahasiswa di salah satu universitas di kota palu. Namanya adalah Universitas Tadulako. Saya angkatan 2022. Hobi saya main game"
```

```python
text = text.lower()
```

```python
tokens = word_tokenize(text)

tokens = [token for token in tokens if token not in string.punctuation]

tokens = [token for token in tokens if not token.isnumeric()]
print('[{}]'.format(', \n'.join(tokens)))
```

```python
stopwords = nltk.corpus.stopwords.words('indonesian')
tokens = [token for token in tokens if token not in stopwords]
print('[{}]'.format(', \n'.join(tokens)))
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()

stemmed_tokens = [stemmer.stem(token) for token in tokens]
print('[{}]'.format(', \n'.join(stemmed_tokens)))
```

```python
processed_text = ' '.join(stemmed_tokens)

print(processed_text)
```

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/valorant.jpeg', 1)
```

```python
resized_img = cv2.resize(img, (400, 225))
cv2_imshow(resized_img)
```

```python
flipped_horizontal = cv2.flip(resized_img, 1)
cv2_imshow(flipped_horizontal)
```

```python
flipped_vertical = cv2.flip(resized_img, 0)
cv2_imshow(flipped_vertical)
```

```python
rotated = cv2.rotate(resized_img, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated)
```

```python
blurred = cv2.blur(resized_img, (10,10))
cv2_imshow(blurred)
```
