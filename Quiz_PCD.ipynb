```python
# Import Library
import cv2
import PIL
import numpy as np
import random

from google.colab.patches import cv2_imshow
from PIL import Image
```

# Baca gambar menggunakan opencv dan PIL

```python
path = ('/content/plat.jpeg')

#Opencv
img_opencv = cv2.imread(path)
img_opencv = cv2.resize(img_opencv, (400, 225))
cv2_imshow(img_opencv)

# PIL
img_pil = Image.open(path)
img_pil = img_pil.resize((400, 225))
img_np = np.array(img_pil)
cv2_imshow(img_np)
```

# Lakukan edge detection menggunakan sobel dan canny



```python
# Sobel OpenCV
sobel_x_opencv = cv2.Sobel(img_opencv, cv2.CV_64F, 1, 0, ksize=3)
sobel_y_opencv = cv2.Sobel(img_opencv, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined_opencv = cv2.magnitude(sobel_x_opencv, sobel_y_opencv)
cv2_imshow(sobel_combined_opencv)
```

```python
#Canny OpenCV
canny_opencv = cv2.Canny(img_opencv, 100, 200)
cv2_imshow(canny_opencv)
```

# Lakukan random cropping dan half-size center cropping

---




```python
# Random cropping
random_height, random_width = 200, 200
height, width = img_opencv.shape[:2]
random_y = np.random.randint(0, height - random_height)
random_x = np.random.randint(0, width - random_width)
randomcrop = img_opencv[random_y:random_y+random_height, random_x:random_x+random_width]
cv2_imshow(randomcrop)
```

```python
# Half size center cropping
center_y = height // 2
center_x = width // 2

half_size_center_crop = img_opencv[center_y - height // 4 : center_y + height // 4,
                                    center_x - width // 4 : center_x + width // 4]

cv2_imshow(half_size_center_crop)
```

# Lakukan transformasi pada gambar : putih jadi hitam, hitam jadi putih

```python
inverted_img = cv2.bitwise_not(img_opencv)

cv2_imshow(inverted_img)
```

# Tambahkan noise pada gambar original : salt and pepper noise dan gaussian

```python
# Salt and Pepper
salt_pepper_amount = 0.1
salt_pepper_noise = np.random.rand(*img_opencv.shape) < salt_pepper_amount
img_saltpepper = np.copy(img_opencv)
img_saltpepper[salt_pepper_noise] = 255

cv2_imshow(img_saltpepper)
```

```python
# Gaussian
mean = 0
std_dev = 10
gaussian_noise = np.random.normal(mean, std_dev, img_opencv.shape)
img_gaussian = cv2.add(img_opencv, gaussian_noise.astype(np.uint8))
cv2_imshow(img_gaussian)
```

# Lakukan transformasi geometri : rotasi, flip, vertikal, dan horizontal


```python
# Rotasi
rotated = cv2.rotate(img_opencv, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated)
```

```python
# Flip vertikal
flipped_vertical = cv2.flip(img_opencv, 0)
cv2_imshow(flipped_vertical)
```

```python
# Flip horizontal
flipped_horizontal = cv2.flip(img_opencv, 1)
cv2_imshow(flipped_horizontal)
```
